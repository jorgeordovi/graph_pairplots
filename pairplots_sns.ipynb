```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = sns.load_dataset('iris')
```

```python

iris
```

```python
sns.pairplot(data=iris, hue='species', diag_kind='hist', markers=["o", "s", "D"])

plt.show()
```

```python
sns.pairplot(data=iris, hue='species', diag_kind='hist')
```

```python
sns.pairplot(data=iris, hue='species', diag_kind='kde')
```

```python
sns.pairplot(data=iris, hue='species', kind='reg')
```

```python
numeric_columns = iris.select_dtypes(include=['Float64', 'int64'])

correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=.5, linecolor='black', cbar_kws={'label': 'Correlation'})

plt.title('Correlation Matrix Iris Dataset')
plt.xlabel('Caracteristics')
plt.ylabel('More Caracterisits')

plt.show()
```

```python
species = iris.pop('species')
sns.clustermap(iris)
```

```python
sns.clustermap(iris, cmap='mako', vmin=0, vmax=10)
```

```python
sns.clustermap(iris, metric='correlation', method='single')
```

```python
sns.clustermap(
    iris,
    figsize=(7,5),
    row_cluster=False,
    dendrogram_ratio=(.1,.2),
    cbar_pos=(0,.2,.03,.4)
)
```

```python
lut = dict(zip(species.unique(), "rgb"))
row_colors=species.map(lut)
sns.clustermap(iris, row_colors=row_colors)
```

```python
sns.clustermap(iris, standard_scale=1)
```

```python
sns.clustermap(iris, z_score=0,  cmap="vlag", center=0)
```

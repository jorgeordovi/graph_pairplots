```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = sns.load_dataset('iris')
```

```python

iris
```

```python
sns.pairplot(data=iris, hue='species', diag_kind='hist', markers=["o", "s", "D"])

plt.show()
```

```python
sns.pairplot(data=iris, hue='species', diag_kind='hist')
```

```python
sns.pairplot(data=iris, hue='species', diag_kind='kde')
```

```python
sns.pairplot(data=iris, hue='species', kind='reg')
```

```python
numeric_columns = iris.select_dtypes(include=['Float64', 'int64'])

correlation_matrix = numeric_columns.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=.5, linecolor='black', cbar_kws={'label': 'Correlation'})

plt.title('Correlation Matrix Iris Dataset')
plt.xlabel('Caracteristics')
plt.ylabel('More Caracterisits')

plt.show()
```
